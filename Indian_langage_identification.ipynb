```
def tokenize_word(path,lang):
  toc = []
  with open(path) as s:
    for i in s:
      toc.append(tokenize(i, lang))
  return toc
```

```
def character_split(lang):
  final_character_set = []
  for m in lang:
    test =[]
    for a in range(0,len(m)):
      test.append(m[a]) 
    final_character_set.append(test)
  return final_character_set
```

```
seta = character_split(langage)
final = []
for i in seta:
  k = []
  for j in i:
      k.append(j)
  final.append(k)
len(final)
```

```
#word2vec implimentation
num_features = 100    # Word vector dimensionality                      
min_word_count = 1   # Minimum word count                        
num_workers = 4       # Number of threads to run in parallel
context = 10          # Context window size                                                                                    
downsampling = 1e-3   # Downsample setting for frequent words

from gensim.models import word2vec
print ("Training model...")
model = word2vec.Word2Vec(final, workers=num_workers, 
                          size=num_features, min_count = min_word_count, 
                          window = context, sample = downsampling)
print ("Finish training model...")
```

```
#def Lstm for identfication of langa
from keras.layers import Embedding, LSTM, Dense, Dropout, Lambda, Flatten
from keras.models import Sequential, load_model, model_from_config
import keras.backend as K

def get_model():
    """Define the model."""
    model = Sequential()
    model.add(LSTM(500, dropout=0.4, recurrent_dropout=0.4, input_shape=[1,100], return_sequences=True))
    model.add(LSTM(360, recurrent_dropout=0.4,activation='relu',return_sequences=True))
    model.add(LSTM(240, recurrent_dropout=0.4,activation='relu',return_sequences=True))
    model.add(LSTM(128, recurrent_dropout=0.4))
    model.add(Dropout(0.5))
    model.add(Dense(5, activation='softmax'))


    model.compile(loss='sparse_categorical_crossentropy', optimizer='Nadam',metrics=['sparse_categorical_accuracy'])
    model.summary()

    return model
```

```
def makeFeatureVec(char, model, num_features):
    """Make Feature Vector from the words list of an Essay."""
    featureVec = np.zeros((num_features,),dtype="float32")
    num_words = 0.
    index2word_set = set(model.wv.index2word)
    for c in char:
        if c in index2word_set:
            num_words += 1
            featureVec = np.add(featureVec,model[c])        
    featureVec = np.divide(featureVec,num_words)
    return featureVec

def getAvgFeatureVecs(char, model, num_features):
    """Main function to generate the word vectors for word2vec model."""
    counter = 0
    FeatureVecs = np.zeros((len(char),num_features),dtype="float32")
    for c in char:
        FeatureVecs[counter] = makeFeatureVec(c, model, num_features)
        counter = counter + 1
    return FeatureVecs
```

```
import numpy as np
langage = getAvgFeatureVecs(final, model, num_features)
langage = np.reshape(langage, (langage.shape[0], 1, langage.shape[1]))
```

```
epochs = 20
batch_size = 64
Lstm = get_model()
history = Lstm.fit(X_train, y_train, epochs=epochs,callbacks=[EarlyStopping(monitor='val_loss', patience=3)])
```
